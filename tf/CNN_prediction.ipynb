```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import pickle
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
im_path = 'C:/Users/19083/code/jupyter/datasets/mine/temp/cattydog.jpg'
im_size = 50
class_names = ['cat', 'dog']
im = cv2.imread(im_path)
plt.imshow(im)
```

```python
model = tf.keras.models.load_model('catty_dog_watcher.model')
img = tf.keras.preprocessing.image.load_img(
    im_path, target_size=(im_size, im_size), color_mode="grayscale"
)
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch

predictions = model.predict(img_array)
score = tf.nn.softmax(predictions[0])

print(str(np.max(score)*100)+'% confidence to be '+class_names[np.argmax(score)])
```
